### Amaliyot. Simple Linear Regression.



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/drive/MyDrive/praktikum_datasets/reklama_sales.csv')
df.head()
```

```python
df.describe()
```

```python
df.corrwith(df['Sales'])
```

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(df, test_size=0.1, random_state=42)

from sklearn import linear_model
LR_model = linear_model.LinearRegression()

x_train = train_set['TV'].to_numpy()
y_train = train_set['Sales'].to_numpy()

#Model train
LR_model.fit(x_train.reshape(-1, 1), y_train)

#theta0 va theta1 koeffitsiyentlarini ajratib olamiz
theta1 = LR_model.coef_[0]   # Accessing the first element of the coefficients array
theta0 = LR_model.intercept_  # Directly accessing the intercept since it is a scalar
print(f'theta0 = {theta0}')
print(f'theta1 = {theta1}')
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='TV', y='Sales', label='Training Data')
plt.plot(x_train, theta1*x_train+theta0, '-r')
plt.xlabel('TV')
plt.ylabel('Sales')
plt.title('Linear Regression')
plt.legend()
plt.show()
```

```python
x_test = test_set['TV'].to_numpy()
y_test = test_set['Sales'].to_numpy()
```

```python
y_pred = LR_model.predict(x_test.reshape(-1, 1))

from sklearn.metrics import mean_absolute_error, mean_squared_error
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print(f'MAE = {mae}')
print(f'RMSE = {rmse}')
```
