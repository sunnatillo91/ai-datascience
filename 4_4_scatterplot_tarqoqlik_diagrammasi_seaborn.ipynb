## 4.4 Scatter plot - Tarqoqlik diagrammasi



```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns  # agar seaborn o'rnatilmagan bo'lsa pip install seaborn
```

```python
df = pd.read_csv('/content/drive/MyDrive/praktikum_datasets/merc.csv')
# df.columns = ['maydon', 'aholi']
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='year', y='price')
plt.show()
```

Sifatliroq ma'lumot olish uchun, 2000 yildan oldin chiqqan mashinalarni tashlab yuboramiz

```python
df = df[df.year>2000]
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='year', y='price')
plt.show()
```

```python
df = df[df.year>2000]
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='mileage', y='price')
plt.show()
```

# Scatterplot 3 va undan ortiq qiymat berish

```python
df = df[df.year>2000]
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='mileage', y='price', hue='transmission')
plt.show()
```

```python
df = df[df.year>2000]
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='mileage', y='price', hue='transmission', style='fuelType', s=100)
plt.show()
```

```python
df = df[df.year>2000]
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='mileage', y='price', hue='fuelType', s=100)
plt.show()
```

# Regplot

```python
df = df[df.year>2000]
plt.figure(figsize=(10,6))
sns.regplot(data=df, x='mileage', y='price')
plt.show()
```

# Swarm plot and strip plot

```python
sns.stripplot(data=df, x=df['transmission'], y=df['price'])
plt.show()
```

```python
sns.swarmplot(data=df, x=df['transmission'], y=df['price'])
plt.show()
```
