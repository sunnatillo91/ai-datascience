```python
import pandas as pd
import sqlite3 as sql
import pprint as print
```

```python
db = sql.connect('/content/drive/MyDrive/praktikum_datasets/electricity.db')
cursor = db.cursor()
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
cursor.fetchall()
```

```python
cursor.execute("PRAGMA table_info(elect);")
cursor.fetchall()
```

```python
pd.read_sql("SELECT * FROM elect LIMIT 10", db, index_col="index")
```

```python
query = "SELECT DISTINCT country FROM elect"
countries = pd.read_sql(query, db)
```

```python
'Uzbekistan' in countries['country'].values
```

```python
query = "SELECT country, year, coal, gas, hydro, oil FROM elect WHERE country='Uzbekistan' AND year>2011"
dfuz = pd.read_sql(query,db)
dfuz
```
