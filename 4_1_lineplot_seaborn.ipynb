```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns  # agar seaborn o'rnatilmagan bo'lsa pip install seaborn
```

```python
df = pd.read_csv('/content/drive/MyDrive/praktikum_datasets/museum_visitors.csv', index_col = 0)
df.head()
```

```python
sns.lineplot(data=df)
plt.show()
```

```python
plt.plot(df)
plt.show()
```

```python
plt.figure(figsize=(16,6))
plt.title("Muzeylarga tashriflar soni")
plt.xlabel("Sana")
plt.ylabel("Odamlar soni")
plt.xticks(rotation=90)  # x o'qidagi qiymatlar 90 gradusga aylantirildi
sns.lineplot(data=df[['Avila Adobe', 'Chinese American Museum']])
plt.show()
```
