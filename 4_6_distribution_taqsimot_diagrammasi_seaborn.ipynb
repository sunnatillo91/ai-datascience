## 4.6 Distribution - Taqsimot diagrammasi



```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns  # agar seaborn o'rnatilmagan bo'lsa pip install seaborn
```

```python
df = pd.read_csv('/content/drive/MyDrive/praktikum_datasets/merc.csv')
df = df[df.year>=2010] # 2010 - yildan keyingi mashinalarni saqlab qolamiz
df = df[df.transmission !='Other']  # avtomat, mexanika, yarim avtomat mashinalarni saqlab qolamiz
df = df[df.fuelType != 'Other']  # benzin, dizel, gibrid mashinalarni saqlab qolamiz
df.head()
```

# Histplot

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='price')
plt.show()
```

# Displot

```python
sns.displot(data=df, x='price', hue='transmission', height=5, aspect=2)
plt.show()
```

```python
sns.displot(data=df, x='price', hue='transmission', multiple='stack', height=5, aspect=2)
plt.show()
```

```python
sns.displot(data=df, x='price', col='transmission', height=5, aspect=2)
plt.show()
```
