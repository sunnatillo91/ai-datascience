```python
db = sql.connect('drive/MyDrive/si_loyihalar/chinook.db')
# db = sql.connect('chinook.db')
cursor = db.cursor()
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")  # Databasedagi barcha jadvallarni chaqirish
cursor.fetchall()
```

```python
cursor.execute('SELECT COUNT (*) FROM customers')
cursor.fetchall()
```

```python
cursor.execute('PRAGMA table_info(customers);')  # Jadvaldagi ustunlarni ko'rish
cursor.fetchall()
```

```python
df = pd.read_sql("SELECT * FROM customers", db, index_col='CustomerId')  # Databasedagi jadvalni dataFrame ko'rinishida chaqirish
df
df = pd.read_sql("SELECT * FROM customers LIMIT 5", db, index_col='CustomerId')  # Databasedagi jadvalni dataFrame ko'rinishida kerakli miqdorda chaqirish
df
```

```python
df = pd.read_sql("SELECT CustomerId, Firstname, Lastname, phone FROM customers", db, index_col='CustomerId')  # Databasedagi jadvalni dataFrame ko'rinishida chaqirish
df.head()
```

```python
query = "SELECT CustomerId, Firstname, Lastname, phone, company FROM customers"
df = pd.read_sql(query, db, index_col='CustomerId')
df.head(3)
```

# Ustun bo'yicha filtr

```python
query = "SELECT CustomerId, Firstname, Lastname, phone, company FROM customers ORDER BY LastName"
df = pd.read_sql(query, db, index_col='CustomerId')
df.head(3)
```

```python
query = "SELECT DISTINCT Country FROM customers"
cursor.execute(query)
cursor.fetchall()
```

```python
df = pd.read_sql(query,db)
df.head()
```
