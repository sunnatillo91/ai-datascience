#Arifmetik amallar

```python
import pandas as pd
import numpy as np
```

```python
s1 = pd.Series([7.3,-2.5,3.4,1.5], index=['a','c','d','e'])
s2 = pd.Series([-2.1,3.6,-1.5,4,3.1], index=['a','c','e','f','g'])
```

```python
s1
```

```python
s2
```

```python
s1+s2
```

```python
df1 = pd.DataFrame(np.arange(9.).reshape(3,3), columns=list('bcd'), index=['Olma', 'Anor', 'Uzum'])
df2 = pd.DataFrame(np.arange(12).reshape(4,3), columns=list('abc'), index=['Olma','Anor', 'Qovun', 'Anjir'])
```

```python
df1
```

```python
df2
```

```python
df1+df2  # ikkala df da ham bir xil qiymatlarnigina qo'shish mumkin
```

```python
df1.add(df2, fill_value=0)  # df lardan birida bo'lib 2-sisda qiymat bo'lmasa unga 0 yoki boshqa raqam berib qo'shish mumkin
```

```python
df2.sub(df1, fill_value=0)   # df lardan birida bo'lib 2-sisda qiymat bo'lmasa unga 0 yoki boshqa raqam berib ayirish mumkin
```

```python
df1.sub(df2, fill_value=0)
```

```python
df2.rsub(df1, fill_value=0)  # rsub() teskari ayirma ya'ni df1 dan df2 ni ayiradi
```

```python
df2.div(df1, fill_value=1)  # df2 ni df1 ga bo'lish
```

```python
df1.rdiv(df2, fill_value=1)  # df2 ni df1 ga teskari bo'lish
```
