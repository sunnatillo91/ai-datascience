# Sodda misollar

```python
A = pd.DataFrame({'Transport':['poyezd', 'avtomobil','kema'],
                  'narxi':['98000', '10000', 65000]})
A
```

```python
B = pd.DataFrame({'Transport':['poyezd', 'avtomobil','tramvay'],
                  'Tezligi':['70', '180', 45]})
B
```

```python
merged = pd.merge(A, B) # inner join - ichki ya'ni ikkalasida ham mavjud xususiyatlargina qaytadi
merged
```

```python
merged = pd.merge(A, B, how='inner') # inner join - ichki ya'ni ikkalasida ham mavjud xususiyatlargina qaytadi
merged
```

# Outer join tashqi bog'lash

```python
merged_outer = pd.merge(A, B, how='outer') # outer join - tashqi ya'ni ikkalasida mavjud bo'lmagan xususiyatlar ham qaytadi
merged_outer
```

```python
A = pd.DataFrame({'Transport_turi':['poyezd', 'avtomobil','kema'],
                  'narxi':['98000', '10000', 65000]})
A
```

```python
B
```

# Ustun nomiga ko'ra bog'lash (merge)

```python
merg = pd.merge(A,B, left_on='Transport_turi', right_on='Transport')
merg
```

# Indekslar ustida birlashtirish

```python
A = A.set_index('Transport_turi')
A
```

```python
B
```

```python
merged1 = pd.merge(A, B, left_index=True, right_on='Transport')
merged1
```

```python
B = B.set_index("Transport")
B
```

```python
merged2 = pd.merge(A, B, left_index=True, right_index=True)  # inner join
merged2
```

```python
merged2 = pd.merge(A, B, left_index=True, right_index=True, how='outer')  # outer join
merged2
```
