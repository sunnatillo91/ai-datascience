```python
df = pd.read_csv('/content/drive/MyDrive/praktikum_datasets/Practicum_Datasets/world_cities_air_water_quality.csv')
df.head()
```

```python
df.columns
```

```python
columns = ['City', "Region", "Country", "AirQuality",
       "WaterPollution"]
df.columns = columns
df.head()
```

```python
df.Region = df.Region.map(lambda x:x.replace('"', ""))
df.head()
```

```python
df.Country = df.Country.map(lambda x:x.replace('"', ""))
df.head()
```

```python
df.to_csv('/content/drive/MyDrive/praktikum_datasets/Practicum_Datasets/updated_world_cities_air_water_quality.csv', index=False)
```

```python
df.Country[2]
```

```python
df.Country = df.Country.map(lambda x:x.strip())  # Ortiqcha bo'sh joy o'chirildi
df.Country[2]
```

```python
df[df.Country == "Germany"].head()
```

```python
df1 = df.copy()
df1.drop('Region', axis=1)
```

# pivot_table orqali irarxik indekslash

```python
df1 = pd.pivot_table(df, index=['Country', 'City'], aggfunc='sum')

print(df1)
```

# Qayta indekslash
- swaplevel()

```python
df1.swaplevel('Country', 'City')
```

```python
df2 = df1.copy()
df2.drop('Region',axis=1, inplace=True)
df2['AirPollution'] = 100-df2['AirQuality'] # AirPollution ustuniga teskari ko'rsatgichli ustun yaratish
df2
df2['WaterQuality'] = 100-df2['WaterPollution'] # WaterQuality ustuniga teskari ko'rsatgichli ustun yaratish
df2
```

```python
df3 = pd.DataFrame(df2.values, index = df2.index, columns= [['AIR', 'AIR', 'WATER', 'WATER'], ['AirQuality', 'AirPollution', 'WaterQuality', 'WaterPollution']]) #
df3
```
