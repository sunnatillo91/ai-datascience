```python
x = np.arange(0,4*np.pi, 0.1)
y = np.sin(x)
```

```python
plt.figure(figsize=(10,5)) # grafik o'lchamlarini o'zgartirish
plt.plot(x,y, linewidth=2, linestyle="-", color='g', marker='o') # grafik chizish va unga turli shakllar berish
plt.title('sin(x) funksiya') # jadvalga nom berish
plt.xlabel('pi')   # x o'qiga nom berish
plt.ylabel('sin(x)')   # y o'qiga nom berish
plt.grid(color='k', linestyle='--')  # orqa fonda kataklar qo'shish
plt.show()
```

# Ikkita jadvalni birga chizish

```python
x = np.arange(0,4*np.pi, 0.1)
sin = np.sin(x)  # sin(x)
cos = np.cos(x)  # cos(x)

plt.figure(figsize=(12,6)) # grafik o'lchamlarini o'zgartirish
# sin(x) grafigi
plt.plot(x, sin,
         linewidth=1.5,
         linestyle="-.",
         alpha=0.7, # chiziqning shaffofligi
         color='g',
         marker='o') # grafik chizish va unga turli shakllar berish
# cos(x) grafigi
plt.plot(x, cos,
         linewidth=1.5,
         linestyle="-.",
         alpha=0.7, # chiziqning shaffofligi
         color='b',
         marker='v') # grafik chizish va unga turli shakllar berish
plt.title('sin(x) va cos(x) funksiya grafigi') # jadvalga nom berish

plt.xlabel('pi')   # x o'qiga nom berish
# Har bir chiziq haqida ma'lumot:
plt.legend(['sin(x)', 'cos(x)'], loc=0, fontsize='x-large')
plt.grid(color='tomato', alpha=0.5, linestyle='--')  # orqa fonda kataklarni o'zgartirish
plt.show()
```
