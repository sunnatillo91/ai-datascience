```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/praktikum_datasets/automobile_data_nan.csv')
df.head()
```

# NaN qiymatlarga belgilangan ustun bo'yicha standart qiymat berish

```python
df.fillna({'horsepower':100}).head()
```

# NaN qiymatlarga belgilangan ustun bo'yicha o'rtacha qiymat berish

```python
df.fillna({'horsepower':np.mean(df['horsepower']), 'wheel-base':np.mean(df['wheel-base'])}).head()
```

```python
np.mean(df['horsepower'])
```

# NaN qiymatlarga o'zidan oldingi qiymatni berish

```python
df.fillna(method='ffill')
```

# NaN qiymatlarga o'zidan keyingi qiymatni berish

```python
df.fillna(method='bfill')
```

# NaN qiymatlarga filtrga ko'ra qiymat berish

```python
df[df['company']=='toyota']
```

```python
df[df.company=='toyota'].fillna(method='bfill')
```
