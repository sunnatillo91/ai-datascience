```python
import pandas as pd
import numpy as np
data = {
    'Boqivoy': {'Triller':5,'Komedya':4, "Drama":1, "Jangari":4, "Serial":0},
    'Oypopuk': {'Triller':3,'Komedya':5, "Drama":5, "Jangari":2, "Serial":3},
    'Solijon': {'Triller':3,'Komedya':4, "Drama":1, "Jangari":5, "Serial":1},
    'Masha': {'Triller':4,'Komedya':5, "Drama":3, "Jangari":1, "Serial":2},
}
df = pd.DataFrame(data)
df
```

```python
# Function to calculate Euclidean distance between two columns
def calculate_euclidean_distance(col1, col2):
    differences = (df[col1] - df[col2]) ** 2
    distance = np.sqrt(differences.sum())
    return distance

# Calculate Euclidean distances between the first column ('Boqivoy') and other columns
distances = {}
for col in df.columns[1:]:
    distances[col] = calculate_euclidean_distance('Boqivoy', col)

distances
```

```python
distances = {}
for col in df.columns[0:3]:
    distances[col] = calculate_euclidean_distance('Masha', col)

distances
```
