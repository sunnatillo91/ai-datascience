## 4.3 Ustunli grafiklar, bar chart


```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns  # agar seaborn o'rnatilmagan bo'lsa pip install seaborn
```

```python
df = pd.read_csv('/content/drive/MyDrive/praktikum_datasets/uzbekistan.csv', index_col = 0)
df.columns = ['maydon', 'aholi']
df.head()
```

```python
plt.figure(figsize=(16,6))
sns.barplot(x=df.index, y=df['aholi'])

plt.title("O'zbekiston respublikasi tarkibidagi hududlar aholi soni")
plt.xlabel("Hudud")
plt.ylabel("Aholi soni (million)")
plt.xticks(rotation=90)  # x o'qidagi qiymatlar 90 gradusga aylantirildi
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.barplot(x=df.index, y=df['maydon'])

plt.title("O'zbekiston respublikasi tarkibidagi hududlar aholi soni")
plt.xlabel("Hudud")
plt.ylabel("Maydoni")
plt.xticks(rotation=90)  # x o'qidagi qiymatlar 90 gradusga aylantirildi
plt.grid()
plt.show()
```

# Jadvaldagi indeksni date formatiga o'tkazish  parse_dates=[0]

```python
df2 = pd.read_csv('/content/drive/MyDrive/praktikum_datasets/museum_visitors.csv', index_col = 0, usecols=[0,1],  parse_dates=[0])
df2.head()
```

```python
type(df2.index)
```

```python
df2['month'] = [i.month for i in df2.index]
df2.head()
```

```python
df2['year'] = [i.year for i in df2.index]
df2.head()
```

```python
df2 = df2.groupby(['month', 'year']).max()
df2.head()
```

```python
df2.unstack(level=0)
```

```python
# Yuqoridagi jadvalning grafikda ko'rinishi
plt.figure(figsize=(12,6))
sns.heatmap(data=df2.unstack(level=0), linewidth=1, square=True, cmap="YlGnBu")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(data=df2.unstack(level=0), linewidth=1, square=True, cmap="YlGnBu")
x_ticks_label = ['Jan', 'Feb', 'Mar', 'Apr', 'May','Jun','Jul','Aug','Sep', 'Oct','Nov','Dec']
plt.xticks(np.arange(12) + .5, labels=x_ticks_label)
plt.show()
```

#Guruhlangan ustunli grafiklar.


```python
df = pd.read_csv("/content/drive/MyDrive/praktikum_datasets/flight_delays.csv")
df.head()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=df['Month'], y=df['AA'])
plt.title("American Airlines kompaniyasi samolyotlari o'rtacha kechikish vaqti")
plt.ylabel("Minutes")
plt.show()
```

```python
data = pd.melt(df, id_vars=['Month'], value_vars=['AA', 'AS', 'DL', 'WN', 'MQ'])
data.head()
```

```python
sns.catplot(data=data, kind="bar", x="Month", y="value", hue="variable")
plt.show()
```

```python
plt.figure(figsize=(14,7))
sns.heatmap(data=df.set_index('Month'), cmap="Accent", annot=True)
plt.xlabel("Airline")
plt.show()
```
