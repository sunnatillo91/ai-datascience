## 8-VAZIFA. Toshkent shahri uchun, har bir yilda o'rtacha min va max haroratlarni ko'rsatuvchi grafik chizing. **Grafikni faylga saqlab oling**

```python
# JAVOBNI SHUYERGA YOZING
t_max = df['TM']  # temp max
t_min = df['Tm']  # temp min

plt.figure(figsize=(16,8))
# Temp max
plt.plot(df.index, t_max,
         linewidth=1.5,
         linestyle='--',
         color='b',
         alpha=0.7,
         marker='^')
# Temp min
plt.plot(df.index, t_min,
         linewidth=1.5,
         linestyle='--',
         alpha=0.7,
         color='g',
         marker='o')

plt.title('Toshkent shahri uchun yillik o\'rtacha min va max harorat')
plt.xlabel('Yil')
plt.ylabel('Harorat')
plt.legend(["O'rtacha maximum harorat", "O'rtacha minimum harorat"], loc=2, fontsize='x-large')
plt.grid(color='tomato', alpha=0.5, linestyle='--')  # orqa fonda kataklarni o'zgartirish
plt.savefig('tash_temp.png')
plt.show()
```

```python
#@title To'g'ri javob
minT = df['Tm']
maxT = df['TM']
plt.figure(figsize=(16,8))
plt.plot(minT, linewidth=2, linestyle="--", marker='h', markerfacecolor='g', label="O'rtacha minimum harorat")
plt.plot(maxT, linewidth=2, linestyle="-", marker='v', markerfacecolor='r', label="O'rtacha maksimum harorat")
plt.grid()
plt.legend()
plt.title("Toshkent shahri uchun yillik o'rtacha min va max harorat")
plt.xlabel("Yil")
plt.ylabel("Harorat")
plt.savefig('tempMinMax.png')
plt.show()
```
