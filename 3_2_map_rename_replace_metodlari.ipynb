```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/praktikum_datasets/automobile_data_nan.csv')
df.head()
```

```python
df['num-of-cylinders'].unique()
```

# Map() qiymatni moslashtirish

```python
dict = {'four':4, 'six':6, 'five':5, 'three':3, 'twelve':12, 'two':2, 'eight':8}
df['cylinders'] = df['num-of-cylinders'].map(dict)
df.head()
```

```python
kurs = 12600
df['price_uz'] = df['price'].map(lambda x: x*kurs)
df
```

# Replace() metodi qiymatni o'zgartirish

```python
df.replace('four',4)
```

```python
# Bir nechta qiymatni o'zgartirish
df.replace(dict)
```
