```python
# Keyingi mashqlar uchun Githubdan kattaroq DF yuklab olamiz
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/usa_cars.csv', index_col=0)
df.head()
```

```python
df.sort_index()
```

```python
df.sort_index(axis=1, inplace=True)  # ustun bo'yicha tartiblash va df ga saqlash
```

```python
df.head()
```

```python
df.sort_index(ascending=False)  # tartib raqamini teskari tartibladi
```

```python
df.sort_index(axis=1, ascending=False)  # ustunlar bo'yicha teskari
```

```python
df2 = df.sort_values(by='price')  # narxga ko'ra o'sish bo'yicha tartiblash
df2.head(4)
```

```python
df2 = df.sort_values(by='price', ascending=False)  # narxga ko'ra kamayish bo'yicha tartiblash
df2.head(10)
```

```python
df2 = df.sort_values(by=['price','mileage'], ascending=False)  # narxi va kilometriga ko'ra kamayish bo'yicha tartiblash
df2.head()
```

```python
df.sort_values(by='year')  # yiliga ko'ra tartiblash
```

```python
df['rating'] = df['year'].rank()  # rating degan ustun qo'shildi va yili bo'yicha baholandi
```

```python
df.sort_values('year')
```

```python
df['rating'] = df['price'].rank(method='min')  # rating degan ustun qo'shildi va narxi bo'yicha baholandi
```

```python
df.sort_values('price')
```

```python
df['rating'] = df['mileage'].rank(method = 'min')  # rating degan ustun qo'shildi va kilometri bo'yicha baholandi
```

```python
df.sort_values('mileage')
```
