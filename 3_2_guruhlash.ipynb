```python
import pandas as pd
import numpy as np
```

```python
bins = [0, 1000, 10000, 30000, 50000, 100000, 300000, 1500000]
```

```python
population = df.pop2021
population
```

# cut() metodi yordamida guruhlash

```python
guruh = pd.cut(population, bins)
guruh
```

```python
guruh.value_counts()
```

```python
df['guruh'] = guruh
df.head()
```

```python
df3 = pd.read_csv('/content/drive/MyDrive/praktikum_datasets/automobile_data.csv', index_col=0)
df3.head()
```

```python
df3.price.describe()
```

```python
bins = [0, 10000, 20000, 50000]
```

```python
range = pd.cut(df.price, bins)
range
```

```python
range.value_counts()
```

# Guruhlash

```python
groups = ["arzon", "o'rta", "qimmat"]
range = pd.cut(df3.price, bins, labels=groups)
range.value_counts()
```

# Bir nechta oraliqqa avtomatik guruhlash

```python
range = pd.cut(df3.price, 4)
range.value_counts()
```

# Nisbatan teng miqdordagi guruhlarga ajratish

```python
range = pd.qcut(df3.price, 4)
range.value_counts()
```

# groupby() yordamida barcha ustunlar bo'yicha guruhlarga ajratish

```python
df3.groupby('company').mean()
```

# groupby() yordamida faqat kerakli ustunlar bo'yicha guruhlarga ajratish

```python
df3.groupby('company').price.max()
```

# groupby() yordamida bir necha ustunlar bo'yicha guruhlarga ajratish

```python
df3.groupby(['company', 'body-style']).mean()
```

# Barcha ustunlarni bir necha ustunlar bo'yicha guruhlarga ajratish

```python
df3.groupby(['company', 'body-style']).mean().head()
```

# .agg() - Bir necha funksiyalar yordamida barcha ustunlarni guruhlash

```python
df3.groupby('company').agg([min, max, np.mean])
```

# .agg() - Bir necha funksiyalar yordamida kerakli ustunlarni guruhlash

```python
df3.groupby(['company', 'body-style']).price.agg([min, max, lambda x: np.median(x)])
```

```python
df3.groupby('company').agg([len, min, max]) #len - guruhdagi elementlar soni, min - guruh uchun minimum, max - guruh uchun maximum
```
