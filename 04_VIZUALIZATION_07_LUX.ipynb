# Taqsimot CDF

```python
sns.displot(data=df, x='price', kind='ecdf', height=4, aspect=2)
plt.xticks(rotation=90)
plt.grid(which='both')  #setka (major, minor, both) qiymatlarini berish mumkin
plt.minorticks_on()
plt.show()
```

```python
sns.displot(data=df, x='mileage', kind='ecdf', height=4, aspect=2)
plt.xticks(rotation=90)
plt.grid(which='both')  #setka (major, minor, both) qiymatlarini berish mumkin
plt.minorticks_on()
plt.show()
```
